```python
o = vector(RDF,[0,0])

def u(phi):
    return vector(RDF,[cos(phi),sin(phi)])

def v(phi):
    return vector(RDF,[-sin(phi),cos(phi)])

```

```python
BASE = [arrow(o,u(2*pi*t/100))+arrow(o,v(2*pi*t/100))+circle((0,0),1.01,color='red') for t in range(101)]
animate(BASE,aspect_ratio=1,axes=False)
```
